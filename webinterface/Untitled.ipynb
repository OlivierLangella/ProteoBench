```python
import matplotlib.pyplot as plt
from adjustText import adjust_text
import numpy as np

# Example scatter data
np.random.seed(0)
x = np.random.rand(300)
y = np.random.rand(300)
labels = [f'Point {i}' for i in range(len(x))]

fig, ax = plt.subplots(figsize=(14,10))
ax.scatter(x, y)

texts = []
for xi, yi, label in zip(x, y, labels):
    texts.append(ax.text(xi, yi, label, fontsize=9))

# Adjust labels to prevent overlap
adjust_text(texts, 
            only_move={'points': 'y', 'texts': 'xy'},  # controls label and point movements
            force_points=3.0,
            force_text=2.0,
            expand_points=(1.5, 1.5), 
            expand_text=(1.2, 1.2),
            arrowprops=dict(arrowstyle="->", color='gray', lw=0.5))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Adjusted Labels')
plt.tight_layout()
plt.show()
```
